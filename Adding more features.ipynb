```python
power = pd.DataFrame.from_dict(power).T
power.rename(columns = {"House": "house_party", "Senate":"senate_party", "President":"president_party"}, inplace = True)
power
```

# Merge new data with df

```python
df = df.merge(power, right_index = True, left_on = "congress")
```

# Sponsor in minority vs majority

```python
def sponsor_maj(title, sponsor_party, house, senate):
    if title == "Sen": 
        if sponsor_party == senate:
            return 1
        else:
            return 0
    else:
        if sponsor_party == house:
            return 1
        else:
            return 0
```

```python
df['sponsor_majority'] =  df.apply(lambda x: sponsor_maj(x.title, x.sponsor_party, x.house_party, x.senate_party), axis = 1)
```

# Split congress between parties?

```python
def split_congress(senate, house):
    if senate == house:
        return 1
    else:
        return 0
```

```python
df["united_congress"] =  df.apply(lambda x :split_congress(x.senate_party, x.house_party), axis = 1)
```

# Sponsor and president same party

```python
def pres_party(sponsor_party, president_party):
    if sponsor_party == president_party:
        return 1
    else:
        return 0
```

```python
df["sponsor_pres_party_match"] =  df.apply(lambda x :pres_party(x.sponsor_party, x.president_party), axis = 1)
```
